# Images

```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
```

```python
image = np.random.random((100, 100))
```

```python
plt.imshow(image)
plt.colorbar()
```

```python
plt.imshow(image, cmap='gray', interpolation='nearest')
```

```python
image = np.zeros((100, 100, 3))
```

```python
plt.imshow(image)
```

```python
image[:, :, 0] = 0
image[:20, :20, 0] = 1

image[10:30, 10:30, 1] = 1

plt.imshow(image)
```
